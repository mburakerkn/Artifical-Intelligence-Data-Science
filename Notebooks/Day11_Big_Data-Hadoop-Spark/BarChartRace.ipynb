```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import bar_chart_race as bcr
pd.set_option("display.max_columns",100)
```

```python
df = pd.read_csv("corona_dat.csv")
```

```python
df.head()
```

```python
df2 = df[["date","China","US","Turkey","Brazil","Spain","Italy","France"]]
```

```python
df2.set_index("date",inplace=True)
```

```python
df2.head()
```

```python
df_total=df2.cumsum(axis=0)
```

```python
df_total.head()
```

```python
bcr.bar_chart_race( df =df_total, filename="covid19.mp4", figsize=(8,6), title="MBurak - Covid-19 CAses by Country")
```
